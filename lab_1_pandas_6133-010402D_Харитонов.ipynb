2.8. работа с пропущенными значениями (если они есть): удаление строк с пропущенными значениями, заполнение пропущенных значений средним значением по колонке. Если пропущенных значений нет — намеренно их "генерируете", прибивая какие-то куски данных в np.nan


```python
# print rows where sex is nan
data.loc[data['sex'].isna()]
```

```python
# drop rows where sex is nan
data = data.dropna(subset=['sex'])
# print rows where sex is nan
data.loc[data['sex'].isna()]
```

```python
nan_rows = [0, 1, 179, 180, 342, 343]
numeric_columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

# set nan values in numeric columns in a few rows
for row in nan_rows:
  random_numeric_columns = random.sample(numeric_columns, 2)
  data.loc[row, random_numeric_columns] = np.nan
```

```python
# print rows where any numeric column is nan
data.loc[data[numeric_columns].isna().any(axis=1)]
```

```python
# replace nan in each numeric column by average value in this column for each species
for species in data['species'].unique():
  rows = data['species'] == species
  data.loc[rows, numeric_columns] = data.loc[rows, numeric_columns].fillna(data.loc[rows, numeric_columns].mean())

# print rows where any numeric column is nan
data.loc[data[numeric_columns].isna().any(axis=1)]
```

```python
# print rows to check values that replaced nan
data.loc[nan_rows]
```
