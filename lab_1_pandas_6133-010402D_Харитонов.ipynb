3 Продемонстрировать работу .groupby, на основе группировок в groupby вычисляете агрегатные функции по одной или нескольким колонкам. Какую информацию о данных вы получили, выполнив данную операцию? Запишите ответ в комментариях.



```python
data_copy = data.copy()
data_copy['score'] = data_copy['score'].fillna(0)

# average university score by country 
data_copy[['country', 'score']].groupby('country').mean().sort_values('score', ascending=False).rename(columns={'score': 'average score'})
```

```python
# highest university score by country 
data_copy[['country', 'score']].groupby('country').max().sort_values('score', ascending=False).rename(columns={'score': 'highest score'})
```
